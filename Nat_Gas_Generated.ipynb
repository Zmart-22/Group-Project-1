```python
import pandas as pd 
import json 
import requests
from api_key import api_key

#api_key= ''

base_url= f'https://api.eia.gov/v2/electricity/electric-power-operational-data/data/?frequency=annual&data[0]=generation&facets[fueltypeid][]=NG&facets[location][]=CO&facets[sectorid][]=99&start=2015&end=2023&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000&api_key={api_key}'

response=requests.get(base_url).json()

print(json.dumps(response, indent =4, sort_keys=True))
```

```python
#lists to store the desired values

fueltype = []
gen = []
genunits = []
period = []

#populate the lists

for item in range(len(response['response']['data'])):
    
    fueltype.append(response['response']['data'][item]['fuelTypeDescription'])
    gen.append(response['response']['data'][item]['generation'])
    genunits.append(response['response']['data'][item]['generation-units'])
    period.append(response['response']['data'][item]['period'])

```

```python
#create dataframe from these lists

NatGasGen = {
            'Fuel Type':fueltype,
            'Amount Generated': gen,
            'Units Measured':genunits,
            'Period': period
    
            }

NatGasGen_df = pd.DataFrame(NatGasGen)  
NatGasGen_df
```

```python
#df sorted by year

NatGasGen_df = NatGasGen_df.sort_values(by='Period', ascending=True)
NatGasGen_df = NatGasGen_df.reset_index(drop=True)
NatGasGen_df
```

```python
#Write to CSV

NatGasGen_df.to_csv('Nat_Gas_Generated.csv', index=False)
```
