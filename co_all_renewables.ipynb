```python
import json
import requests
import pandas as pd
import matplotlib.pyplot as plt

from api_key import api_key
```

```python
#Colorado all renewable use
base_url = f'https://api.eia.gov/v2/electricity/electric-power-operational-data/data/?frequency=annual&data[0]=generation&facets[fueltypeid][]=AOR&facets[location][]=CO&facets[sectorid][]=99&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000&api_key={api_key}'
response = requests.get(base_url).json()

#print results of the api request
print(json.dumps(response, indent = 4, sort_keys=True))
```

```python
#make some lists
energy_type = []
period = []
value = []
unit = []
```

```python
#populate the lists
for item in range(len(response['response']['data'])):
    energy_type.append(response['response']['data'][item]['fuelTypeDescription'])
    period.append(response['response']['data'][item]['period'])
    value.append(response['response']['data'][item]['generation'])
    unit.append(response['response']['data'][item]['generation-units'])
```

```python
#create dataframe from these lists
energy = {
    'Period':period,
    'Energy Type': energy_type,
    'Value':value,
    'Unit': unit
    
}

energy_df = pd.DataFrame(energy)
```

```python
#df sorted by year
energy_df = energy_df.sort_values(by='Period', ascending=True)
energy_df = energy_df.reset_index(drop=True)
energy_df
```
