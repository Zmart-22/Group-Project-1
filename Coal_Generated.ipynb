```python
import pandas as pd 
import json 
import requests


api_key= 'EhLybGenW3vz7jb4ocYVLOpGgceGcGrDceGIKNcE'

base_url= f'https://api.eia.gov/v2/electricity/electric-power-operational-data/data/?frequency=annual&data[0]=generation&facets[fueltypeid][]=COW&facets[location][]=CO&facets[sectorid][]=99&start=2015&end=2023&sort[0][column]=period&sort[0][direction]=desc&offset=0&length=5000&api_key={api_key}'

response=requests.get(base_url).json()

print(json.dumps(response, indent =4, sort_keys=True))
```

```python
#lists to store the desired values
fueltype = []
gen = []
genunits = []
period = []

#populate the lists

for item in range(len(response['response']['data'])):
    
    fueltype.append(response['response']['data'][item]['fuelTypeDescription'])
    gen.append(response['response']['data'][item]['generation'])
    genunits.append(response['response']['data'][item]['generation-units'])
    period.append(response['response']['data'][item]['period'])
    
```

```python
#create dataframe from these lists

CoalGen = {
            'Fuel Type':fueltype,
            'Amount Generated': gen,
            'Units Measured':genunits,
            'Period': period
    
            }

CoalGen_df = pd.DataFrame(CoalGen)  
CoalGen_df
```

```python
#df sorted by year

CoalGen_df = CoalGen_df.sort_values(by='Period', ascending=True)
CoalGen_df = CoalGen_df.reset_index(drop=True)
CoalGen_df
```

```python
#Write to CSV

CoalGen_df.to_csv('Coal_Generated.csv', index=False)
```
